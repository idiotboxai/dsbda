```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic-dataset.csv')
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.displot(df['Age']) 
plt.show()
```

```python
plt.figure(figsize=(5, 3))
bp = sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df, palette='pastel', dodge=False)
plt.legend([], [], frameon=False) 
plt.show()
```

```python
plt.figure(figsize=(5, 3))
vp = sns.violinplot(x='Pclass', y='Age', hue='Pclass', data=df, palette='rainbow', dodge=False)
plt.legend([], [], frameon=False) 
plt.show()
```

```python
plt.figure(figsize=(5, 4))
pq = sns.histplot(x='Fare', bins=10, data=df, hue='Survived', kde=False)
for i in pq.containers:
    pq.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(5, 4))
pq = sns.histplot(x='Fare', bins=10, data=df, hue='Survived', kde=False)
for i in pq.containers:
    pq.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(5, 4))
st = sns.scatterplot(x='Age', y='Fare', data=df)
plt.show()
```

```python
plt.figure(figsize=(5, 4))
kl = sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived')
plt.show()
```
