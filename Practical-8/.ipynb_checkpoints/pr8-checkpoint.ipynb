```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset('titanic')

df.head()
```

```python
plt.figure(figsize=(6,4))
sns.displot(df['age']) 
plt.show()
```

```python
plt.figure(figsize=(5, 3))
bp = sns.boxplot(x='class', y='age', hue='class', data=df, palette='pastel', dodge=False)
plt.legend([],[], frameon=False) 
plt.show()
```

```python
plt.figure(figsize=(5, 3))
vp = sns.violinplot(x='class', y='age', hue='class', data=df, palette='rainbow', dodge=False)
plt.legend([], [], frameon=False) 
plt.show()
```

```python
plt.figure(figsize=(5,4))
pq = sns.histplot(x='fare',bins=10,data=df,hue='survived',kde=False)
for i in pq.containers:
  pq.bar_label(i)
plt.show()
```
