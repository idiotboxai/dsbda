```python
df.groupby("Species").describe().sum()
```

```python
df.drop(columns=['Id'], inplace=True)
```

```python
grouped_stats = df.groupby('Species').agg(['mean', 'median', 'min', 'max', 'std'])
print("Grouped Summary Statistics:\n")
print(grouped_stats)
```

```python
species_counts = df['Species'].value_counts().to_dict()
print("\nNumber of samples per species (as a list):")
species_count_list = list(species_counts.values())
print(species_count_list)
```

```python
print("\nDescriptive statistics for each species:")
for species in df['Species'].unique():
    print(f"\nStatistics for {species}:")
    print(df[df['Species'] == species].describe(percentiles=[.25, .5, .75]))
```
